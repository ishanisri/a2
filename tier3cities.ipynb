```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('C:/Users/Harini/Desktop/file1.csv')
table=pd.pivot_table(data,index=['Outlet_Location_Type','Item_Identifier','Item_Type','Outlet_Type'],values=['Item_Type','Item_MRP','Item_Outlet_Sales'])
```

```python
tier3=table.xs('Tier 3',level=0)
tier3
```

```python
table3=pd.pivot_table(tier3,index=['Outlet_Type','Item_Type'],values=['Item_Outlet_Sales'],aggfunc=[np.sum,len])
table3
```

```python
groc=table3.xs('Grocery Store',level=0)
groc
```

```python

n1=groc['sum']



n2=n1.as_matrix(['Item_Outlet_Sales'])

c=[]
a=n2.tolist()
for i in range(len(n2)):
    c.append(a[i][0])


n4=data.Item_Type.unique().tolist()
n4.sort()


d=[]
for i in range(len(n2)):
    d.append(i)
    
    
plt.bar(d,c,width=0.3,color='r')
plt.xticks(d,n4,rotation='vertical')
plt.title('Graph of total sales of items bought in each category in tier 3 city grocery store')
plt.grid()
plt.xlabel('Item Type')
plt.ylabel('total sales')
plt.figure(figsize=(20,3))


plt.show()
n3=n1.sort_values('Item_Outlet_Sales')
n3
```

```python

p1=groc['len']
p2=p1.as_matrix(['Item_Outlet_Sales'])
c1=[]
a1=p2.tolist()
for i in range(len(n2)):
    c1.append(a1[i][0])


p4=data.Item_Type.unique().tolist()
p4.sort()


d1=[]
for i in range(len(p2)):
    d1.append(i)
plt.bar(d1,c1,width=0.3,color='r')
plt.xticks(d1,p4,rotation='vertical')


plt.title('Graph of no of items bought in each category in tier 3 city grocery  store')
plt.xlabel('Item Type')
plt.ylabel('no of items')
plt.figure(figsize=(20,3))


plt.show()
p1.sort_values('Item_Outlet_Sales')
```
